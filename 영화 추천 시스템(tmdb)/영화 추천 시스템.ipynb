```python
C = df2['vote_average'].mean()
C
```

```python
# 영화에 대해 평가한 수 (상위 10% 구하기)
m = df2['vote_count'].quantile(0.9)
m
```

```python
# 평가한 수가 1838개 넘는 데이터 구하기
q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
```

```python
q_movies['vote_count'].sort_values()
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v+m) * R) + (m / (m+v) *C)
```

```python
# score 열 생성
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
q_movies.head(3)
```

```python
# score를 기준으로 내림차순
q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title', 'vote_count', 'vote_average', 'score']].head()
```

## 영화의 제목을 입력받으면 코사인 유사도를 통해서 가장 유사도가 높은 상위 10개의 영화 목록 반환

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    # 영화 제목을 통해서 전체 데이터 기준 그 영화의 index값 얻기
    idx = indices[title]
    
    # 코사인 유사도 메트릭스(cosine_sim)에서 idx에 해당하는 데이터를 [idx, 유사도] 형태로 보기
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # 코사인 유사도 기준으로 내림차순 정렬
    sim_scores= sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # 자기 자신을 제외한 10개의 추천 영화 슬라이싱
    sim_scores = sim_scores[1:11]
    
    # 추천 영화 목록 10개의 인덱스 정보 추출
    movie_indices = [i[0] for i in sim_scores]
                     
    # 인덱스 정보를 통해 영화 제목 추출
    return df2['title'].iloc[movie_indices]
```

```python
test_idx = indices['The Dark Knight Rises'] # 영화 제목을 통해서 전체 데이터 기준 그 영화의 index값 얻기
test_idx
```

```python
cosine_sim[3]
```

```python
# 코사인 유사도 메트릭스(cosine_sim)에서 idx에 해당하는 데이터를 [idx, 유사도] 형태로 보기
test_sim_scores = list(enumerate(cosine_sim[3]))
test_sim_scores[:10]
```

```python
''' key 매개변수에 기준으로 정렬할 변수를 입력한다.
    lambda 함수를 통해 test_sim_scores 값인 유사도를 기준으로 정렬시킨다. 
     reverse=True : 내림차순 
'''
# 코사인 유사도 기준으로 내림차순 정렬
test_sim_scores= sorted(test_sim_scores, key=lambda x: x[1], reverse=True)

# 자기 자신을 제외한 10개의 추천 영화 슬라이싱
test_sim_scores[1:11]
```

```python
# 추천 영화 목록 10개의 인덱스 정보 추출
test_movie_indices = [i[0] for i in test_sim_scores[1:11]]
test_movie_indices
```

```python
# 감독 정보 추출하기
def get_director(x):
    # 여러 딕셔너리를 검색함
    for i in x:
        if i['job'] =='Director':
            return i['name']
    return np.nan
```

```python
df2['director'] = df2['crew'].apply(get_director)
df2['director']
```

```python
# 감독정보가 없는 데이터 확인하기
df2[df2['director'].isnull()]
```

```python
# 처음 3개의 데이터 중에서 name에 해당하는 value만 추출
def get_list(x):
    # 데이터 x가 자료형이 list인지 확인하기
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return []
```

```python
features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head()
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(' ', '')) for i in x]
    else :
        if isinstance(x, str):
            return str.lower(x.replace(' ',''))
        else :
            return ''
```

```python
features = ['cast', 'keywords', 'director','genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head()
```

```python
# keywords, cast, director, genres 값 띄어쓰기로 합치기
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
df2['soup']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
count_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
cosine_sim2
```
